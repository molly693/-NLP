```python
alpha = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"
punc = "！？｡。＂＃＄％＆＇（）＊＋，\一－／：；＜＝＞＠［＼］＾＿｀｛｜｝～｟｠｢｣､、〃》「」『』【】〔〕〖〗〘〙〚〛〜〝〞〟〰〾〿–—‘’‛“”„‟…‧﹏."

# Function
def data_cleaning(sentence):
    sentence = preprocess(sentence)
    sentence = stopwords_cut(sentence)    
    return sentence
    
def preprocess(sentence):
    # remove space, alphabet, digit and puctuation
    sentence = sentence.strip()
    sentence = "".join(char for char in sentence if not char.isdigit() and not char in alpha and not char in punc)
    for strange in strange_str:
        sentence = sentence.replace(strange, "")
    for punctuation in string.punctuation:
        sentence = sentence.replace(punctuation, "")
        
    return sentence

def stopwords_cut(sentence):
    # jieba tokenized and remove stopwords
    seglist = jieba.cut(sentence, cut_all=False)
    final = []
    for seg in seglist:
        if seg not in stopwords:
            final.append(seg)
    output = " ".join(list(final))  
    return output
```
